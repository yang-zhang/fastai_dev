## need state; turn into class

```python
class Norm(Transform):
    def __init__(self,m,s): 
        super().__init__()
        self.m,self.s = m,s
    def encodes(self, x:Tensor): return (x-self.m)/self.s

m,s = timg.mean(),timg.std()
f = Norm(m=m,s=s)
nimg = f(timg)
```

```python
imgtup = (mti,mi)
ttup = f(imgtup)
```

```python
@Transform
def flip_img(x:MyTensorImg): return torch.flip(x, 1) if random.random()>0.5 else x
```

```python
ttup = flip_img(f(imgtup))
```

```python
compose
```

```python
pipe = compose(f, flip_img)
ttup = pipe(imgtup)
```

```python
pipe = Pipeline([f, flip_img])
ttup = pipe(imgtup)
```
